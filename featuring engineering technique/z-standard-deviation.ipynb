# Here we use z-Standard deviation technique 

```python
## Standard Deviation 
### It means how far away the datapoint is from their mean value
```

```python
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (10,6) 
```

```python
df = pd.read_csv('D:\Akash\Machine Learning\Data_for_ML\Book9.csv')
df.sample(5)
```

```python
plt.hist(df.height , bins=20 , rwidth=0.8)
plt.xlabel('Height(inches)')
plt.ylabel('Count')
plt.show()
```

```python
df.height.describe()
```

```python
from scipy.stats import norm
import numpy as np

plt.hist(df.height , bins=20 , rwidth=0.8 , density=True)
plt.xlabel('Height(inches)')
plt.ylabel('Count')

# plotting normal distribution curv here
rng = np.arange(df.height.min() , df.height.max() , 0.1)
plt.plot(rng , norm.pdf(rng,df.height.mean(),df.height.std()))
```

## we generally use here z = 3 , i.e.
# 3-standard deviation 

```python
upper_limit = df.height.mean() + 3 * df.height.std()
lower_limit =  df.height.mean() - 3 * df.height.std()

upper_limit , lower_limit
```

### Anything outside this range is an outlier

```python
df[(df.height > upper_limit) | (df.height < lower_limit)]
```

## we have to remove this

```python
df[(df['height'] > lower_limit) & (df['height'] < upper_limit)]
```
