# Logistic Regression to do Binary Classification

## 1. Import all usefull libraries

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

## 2. Import data

```python
df = pd.read_csv('D:\Akash\Machine Learning\Data_for_ML\Book5.csv')
df
```

## 3. plot data

```python
plt.scatter(df.age , df.have_insurance , marker='+' , color='red')
```

## 4. train_test_splitting of dataset

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df[['age']], df.have_insurance, train_size=0.9)
```

## 5. Import our model and fit the training data

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(x_train , y_train)
```

## 6. Testing Efficiency of our model

```python
model.predict(x_test)
```

```python
model.score(x_test , y_test)
```

## 7. Plot sigmoid function graph 

```python
x_range = 80
Xs = [i for i in range(x_range)]
Ys = [model.predict_proba([[value]])[0][1] for value in range(x_range)]

plt.scatter(df.age , df.have_insurance , marker='+' , color='red')
plt.plot(Xs , Ys , color = 'blue')
```
