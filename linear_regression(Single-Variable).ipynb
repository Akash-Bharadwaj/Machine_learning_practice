# Liner regression for single variable

## 1. import all libraries

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import linear_model
```

## 2. import data

```python
df = pd.read_csv("D:\Akash\Machine Learning\Data_for_ML\Book2.csv")
df
```

## 3. To check Linearty of our Data

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$')
plt.scatter(df.area , df.price , color='red',marker='+')
```

## 4. Fitting data in our model

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3300]])
```

## 5. Rechicking of the result

```python
m = reg.coef_
m
```

```python
b = reg.intercept_
b
```

```python
y = m * 3300 + b
y
```

## 6. Plot the prediction line

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$')
plt.scatter(df.area , df.price , color='red',marker='+')
plt.plot(df.area , reg.predict(df[['area']]) , color='blue')
```

## 7. Save Model into a file

### we save model into a file so that we use this file any where , by calling it .

```python
import joblib 
```

```python
joblib.dump(reg , 'model_linear-regg')
```

## 8. Call model for use

```python
reg = joblib.load('model_linear-regg')
```

```python
reg.predict([[3300]])
```
