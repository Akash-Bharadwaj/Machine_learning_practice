# 1. Matplotlib basics

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,3,4,5,7,8,9]
y = [50,53,52,48,44,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,color='green',linewidth=5,linestyle='dotted')
```

# 3. Axis labels, legend ,Grid

```python
days = [1,2,3,4,5,6,7]
max_t = [23,65,12,45,78,54,15]
min_t = [9,13,23,45,32,76,45]
avg_t = [34,54,33,23,12,19,27]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days,max_t , label='max')
plt.plot(days,min_t , label='min')
plt.plot(days,avg_t , label='avg')

# plt.legend(loc="upper right")
plt.legend(loc="best" , shadow =True, fontsize ='small')
plt.grid()
```

# 4. Bar Chart

```python
company = ['l&t' ,'Bel' ,'Grsb' ,'Tata']
revenue = [40 ,90,60 ,100]
profit = [20 ,25,34 ,50]
```

```python
import numpy as np

ypos = np.arange(len(company))
ypos
```

```python
plt.xticks(ypos ,company)
plt.ylabel("revenue(bin)")
plt.xlabel("company")
plt.title("Defence sector")
plt.bar(ypos - 0.2 ,revenue , width = 0.4 , label="revenue")
plt.bar(ypos + 0.2 ,profit , width=0.4 , label="profit")

plt.legend()
```

```python
plt.xticks(ypos ,company)
plt.title("Defence sector")
plt.barh(ypos - 0.2 ,revenue , label="revenue")
plt.barh(ypos + 0.2 ,profit , label="profit")

plt.legend()
```

# 5. Histograms

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],label=['men','women'])
plt.legend()
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar_men,bins=[80,100,125,150],rwidth=0.95,histtype='step')
# plt.hist(blood_sugar_men, bins=[80,100,125,150], rwidth=0.95, orientation='horizontal')
```

# 6. Pie chart

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```
