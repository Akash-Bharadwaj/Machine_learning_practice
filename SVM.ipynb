# Support Vector Machine 

## Here we use it to do classification of iris datasets

## 1. Import all usefull libraries

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

## 2. Import iris datasets

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

## 3. Have a view on our dataset

```python
dir(iris)
```

```python
iris.feature_names
```

```python
iris.target_names
```

## 4. creating a dataframe by taking features of iris datasets

```python
df = pd.DataFrame(iris.data , columns = iris.feature_names)
df.head()
```

### Add a target datasets in our dataframe

```python
df['target'] = iris.target
df.head()
```

### Add a target datasets in our dataframe

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

## 5. Plot the Dataframe 

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df0 = df[df.target == 0]
df1 = df[df.target == 1]
df2 = df[df.target == 2]
```

```python
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
# we use only two parameters to plot a graph
plt.scatter(df0['sepal length (cm)'] , df0['sepal width (cm)'] , color = 'green',marker='+')
plt.scatter(df1['sepal length (cm)'] , df1['sepal width (cm)'] , color = 'blue',marker='.')
```

```python
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
# we use only two parameters to plot a graph
plt.scatter(df0['petal length (cm)'] , df0['petal width (cm)'] , color = 'green',marker='+')
plt.scatter(df1['petal length (cm)'] , df1['petal width (cm)'] , color = 'blue',marker='.')
```

## 6. train test Splitting of our datasets

### Remove unwanted features from our dataset

```python
x = df.drop(['target','flower_name'], axis='columns')
x.head()
```

```python
y = df.target
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

## 7. Importing our model and fit the training data

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(x_train , y_train)
```

## 8. Testing our model

```python
model.score(x_test , y_test)
```
